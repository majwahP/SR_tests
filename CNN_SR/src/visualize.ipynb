```python
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
import dataset 
import torch
```

```python
def tensor_to_image(tensor):
    """Convert a PyTorch tensor (C, H, W) to a NumPy image (H, W, C)."""
    image = tensor.permute(1, 2, 0).numpy()  # Change dimensions from (C, H, W) â†’ (H, W, C)
    image = (image * 255).astype("uint8")  # Convert from [0,1] to [0,255]
    return image
```

```python
def visualize_lr_hr_images():
    # Load dataset with batch_size=4 for easy visualization
    train_loader, _ = dataset.get_dataloaders(batch_size=4)  # Only need training images for visualization

    # Get one batch of LR and HR images
    lr_images, hr_images = next(iter(train_loader))

    # Create a figure with 2 rows: LR images (top), HR images (bottom)
    fig, axes = plt.subplots(2, 4, figsize=(10, 5))

    for i in range(4):  # Loop through 4 images in the batch
        # Low-resolution image (top row)
        lr_img = tensor_to_image(lr_images[i])
        axes[0, i].imshow(lr_img)
        axes[0, i].axis("off")
        axes[0, i].set_title("Low-Res")

        # High-resolution image (bottom row)
        hr_img = tensor_to_image(hr_images[i])
        axes[1, i].imshow(hr_img)
        axes[1, i].axis("off")
        axes[1, i].set_title("High-Res")

    plt.show()
```

```python
visualize_lr_hr_images()
```
